```python
BOLD = '\033[1m'
UNDERLINE = '\033[4m'
END_FORMATTING = '\033[0m'

# Null Hypothesis
#There is no significant difference in the mean number of cycles rented between working days and holidays

# Alternative Hypothesis
# There is a significant difference in the mean number of cycles rented between working days and holidays.
    
# Here I am separating the rows with the working day as wroking day or holiday
workingday_counts = dataset1[dataset1['workingday'] == 'Working']['count']
non_workingday_counts = dataset1[dataset1['workingday'] == 'Holiday']['count']

# here I am performing the Anova Test to find the relation between the working day and count
t_value, p_value = ttest_ind(workingday_counts, non_workingday_counts)

# Here I am Printing the Results of the Test
print(f"{BOLD}{UNDERLINE}2- Sample T-Test to check if Working Day has an effect on the number of electric cycles rented{END_FORMATTING}")
print("T_Statitic:", t_value)
print("P-value:", p_value)

# Here I am Comparing the p value with the alpha to check the null hypotheisis
if p_value < 0.05:
    print("There is a significant relationship between workingday and count")
else:
    print("There is no significant relationship between workingday and count")
```

```python
# Null Hypothesis 
# There is no significant difference in the mean count of rented cycles across different weather conditions.

# Alternative Hypothesis 
# There is a significant difference in the mean count of rented cycles across different weather conditions.

# Here I am separating the rows 
clear_counts = dataset1[dataset1['weather'] == 'Clear']['count']
cloudy_counts = dataset1[dataset1['weather'] == 'Cloudy']['count']
light_snow_counts = dataset1[dataset1['weather'] == 'Light Snow']['count']
heavy_rain_counts = dataset1[dataset1['weather'] == 'Heavy Rain']['count']

# here I am performing the Anova Test to find the relation between the weather conditions and count
a_value, p_value = f_oneway(clear_counts, cloudy_counts, light_snow_counts, heavy_rain_counts)

# Here I am Printing the Results of the Test
print(f"{BOLD}{UNDERLINE}Anova-Test Results for Weather and Count{END_FORMATTING}")
print("A-Statistic: ",a_value)
print("P_value: " , p_value)

# Here I am Comparing the p value with the alpha to check the null hypotheisis
if p_value < 0.05:
    print("There is a significant relationship between weather conditions and the count of rented cycles.")
else:
    print("There is no significant relationship between weather conditions and the count of rented cycles.")
```

```python
# Null Hypothesis 
# There is no significant difference in the mean count of rented cycles across different seasons.

# Alternative Hypothesis
# There is a significant difference in the mean count of rented cycles across different seasons.

# Here I am separating the rows
spring_counts = dataset1[dataset1['season'] == 'spring']['count']
summer_counts = dataset1[dataset1['season'] == 'summer']['count']
fall_counts = dataset1[dataset1['season'] == 'fall']['count']
winter_counts = dataset1[dataset1['season'] == 'winter']['count']

# here I am performing the Anova Test to find the relation between the different seasons and count
a_value, p_value = f_oneway(spring_counts, summer_counts, fall_counts, winter_counts)

# Here I am Printing the Results of the Test
print(f"{BOLD}{UNDERLINE}Anova-Test Results for Season and Count{END_FORMATTING}")
print("A-Statistic: ",a_value)
print("P_value: " , p_value)

# Here I am Comparing the p value with the alpha to check the null hypotheisis
if p_value < 0.05:
    print("There is a significant relationship between season conditions and the count of rented cycles.")
else:
    print("There is no significant relationship between season conditions and the count of rented cycles.")
```

```python
# Null Hypothesis 
# There is no significant association between weather and season.

# Alternative Hypothesis 
# There is a significant association between weather and season.

# Here I am Craeting a cross table to conduct a Chi - Square - Test
contingency_table = pd.crosstab(dataset1['weather'], dataset1['season'])

# Here I am Conducting the Chi Square Test
c_value, p_value, dof, expected = chi2_contingency(contingency_table)

#Here I am printing the values 
print(f"{BOLD}{UNDERLINE}Chi-square Test Results:{END_FORMATTING}")
print("Chi-square Statistic:", c_value)
print("Degree Of Freedom:" , dof)
print("expected:", expected)
print("P-value:", p_value)

# Here I am Comparing the p value with the alpha to check the null hypotheisis
alpha = 0.05
if p_value < alpha:
    print("There is a significant relationship between weather and season.")
else:
    print("There is no significant relationship between weather and season.")
```
